```python
import pickle
import numpy as np
from config import BASE_DIR

'''
routine to fetch vector for a given cid
'''
def cid2vect(cid):
    # load mapping
    infile = open(BASE_DIR+'/cid_vect_dict.pkl', 'rb')
    mapping = pickle.load(infile)
    infile.close()
    print('Vector found for compund '+ str(cid))
    return (mapping[cid]) 
```

```python
cid2vec_1 = cid2vect(9) # cid2vect(cid)
#print(cid2vec_1)
```

```python
'''
routine to fetch cid for a given vector
'''
def vect2cid(vect):
    # load mapping
    infile = open(BASE_DIR+'/cid_vect_dict.pkl', 'rb')
    mapping = pickle.load(infile)
    infile.close()

    queryres = []

    queryvec = np.asarray(vect)

    for key in mapping:
        keyvec = np.asarray((mapping[key]))
        if np.array_equal(keyvec, queryvec):
            queryres = key
            break

    return queryres
```

```python
vec2cid_1 = vect2cid(cid2vec_1)
print(vec2cid_1)
```
